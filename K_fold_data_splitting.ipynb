```python
import pandas as pd
from sklearn.model_selection import KFold
import numpy as np

df = pd.read_csv('data/iris_preprocessed.csv')
print("Shape:", df.shape)
df.head()
```

```python
X = df.drop(columns=['label'])
y = df['label']


kf = KFold(n_splits=5, shuffle=True, random_state=42)

X_train_arr = []
X_test_arr = []
y_train_arr = []
y_test_arr = []

for fold, (train_index, test_index) in enumerate(kf.split(X), 1):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    X_train_arr.append(X_train)
    X_test_arr.append(X_test)
    y_train_arr.append(y_train)
    y_test_arr.append(y_test)

    
    # print(f"Fold {fold}:")
    # print(f"  Train indices: {train_index}")
    # print(f"  Test indices:  {test_index}")

    # Here to place the model training function
```

```python
X_train_arr.to_csv('data/X_train_arr.csv', index=False)
X_test_arr.to_csv('data/X_test_arr.csv', index=False)
y_train_arr.to_csv('data/y_train_arr.csv', index=False)
y_test_arr.to_csv('data/y_test_arr.csv', index=False)
```
