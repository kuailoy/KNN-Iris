```python
# Iris Dataset – Data Preprocessing
'''
Steps covered:
1. Load the dataset and explore features
2. Check and handle missing values
3. Encode categorical variables
4. Apply feature scaling (Standardization & Normalization)
'''
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler

# Load the dataset 
df = pd.read_csv('data/iris_with_missing.csv')

print("Shape:", df.shape)
df.head()

print('\n')

print("First 5 rows of the dataset:\n") 
print(df.head())
```

```python
# Describe Features
print("Data types:\n", df.dtypes)
print("\nClass distribution:\n", df['class'].value_counts())
print("\nStatistical summary:")
df.describe()
```

```python
# Section2: Handle Missing Values

# Identify columns with missing values
df.isna().sum()
```

```python
# Fill numeric missing values with column mean
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill categorical missing values with mode
cat_cols = df.select_dtypes(include=['object', 'string']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
# Encode Categorical Variables
le = LabelEncoder()
df['class_encoded'] = le.fit_transform(df['class'])

print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
print("\nSample (class vs encoded):")
df[['class', 'class_encoded']].drop_duplicates().sort_values('class_encoded')
```

```python
# Feature Scaling
feature_cols = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Standardization (zero mean, unit variance)
scaler_std = StandardScaler()
X_standardized = pd.DataFrame(
    scaler_std.fit_transform(df[feature_cols]),
    columns=[f'{c}_std' for c in feature_cols]
)

# Normalization (scale to [0, 1])
scaler_norm = MinMaxScaler()
X_normalized = pd.DataFrame(
    scaler_norm.fit_transform(df[feature_cols]),
    columns=[f'{c}_norm' for c in feature_cols]
)

print("=== Standardized Features (first 5 rows) ===")
print(X_standardized.head())
print("\n=== Normalized Features (first 5 rows) ===")
print(X_normalized.head())
```

```python
# Final Preprocessed Dataset
df_preprocessed = pd.concat([X_standardized, df['class_encoded']], axis=1)
df_preprocessed.rename(columns={'class_encoded': 'label'}, inplace=True)

print("Preprocessed dataset shape:", df_preprocessed.shape)
df_preprocessed.head()
```

```python
# --- 8. Save Preprocessed Data ---
df_preprocessed.to_csv('data/iris_preprocessed.csv', index=False)
print("Saved to data/iris_preprocessed.csv")
```
